```python
df = pd.concat([pd.read_csv(f) for f in files], keys=files) 
```

```python
#Percentage of missing data.
def missing_data(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
missing_data(df)
```

```python
df= df.dropna()
df
```

```python
mydict = {'tripduration':'Trip Duration',
          'starttime':'Start Date', 
          'stoptime': 'End Date', 
          'start station id': 'Start Station ID',
          'start station name': 'Start Station Name', 
          'start station latitude': 'Start Station Latitude',
          'start station longitude': 'Start Station Longitude', 
          'end station id': 'End Station ID', 
          'end station name': 'End Station Name',
          'end station latitude': 'End Station Latitude', 
          'end station longitude': 'End Station Longitude', 
          'bikeid': 'Bike ID', 'usertype':'User Type',
          'birth year': 'Birth Year', 
          'gender':'Gender'}
```

```python
df = df.rename(columns=mydict)
df
```

```python
df.loc[df.Gender == 0, 'Gender'] = 'Unknown'
df.loc[df.Gender == 1, 'Gender'] = 'Male'
df.loc[df.Gender == 2, 'Gender'] = 'Female'
df = df.drop(df.index[(df['Birth Year'] < 1950)])
df['Age'] = 2020-df['Birth Year']

df['Trip Duration'] = df['Trip Duration'].div(60)

df['Free'] = False
df['Free'] = df.loc[df['Trip Duration'] <= 30, 'Free'] = True

df['Number of Records'] = 1

df = df.drop(df.index[(df['Trip Duration'] < 1.5) & 
                          (df['Start Station Latitude'] == df['End Station Latitude'])])
df = df.drop(df.index[(df['Trip Duration'] > 120)])

df = df.dropna()
```
